```python
# Handling missing values(mean and scikit-lerarn)
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
# Finding Unique values in Country
ds['Country'].unique()
```

```python
# Encoding categorical column (One-Hot Encoding)
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encoded_data=ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# Make Column header and make a dataframe for array
column_names=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data,columns=column_names)
encoded_df
```

```python
# Joint original dataset and new dataframe
ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
# Droping Categorical column
ds=ds.drop(columns=['Country'])
ds
```

```python
# Lable Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
# Feature Scaling(Reducing Larger data values to its smaller data values)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
# Divide dataset as Independent(input) and Dependent(output) variables
x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```
